### c) 
When training the model using the whole dataset, we do not know how well the model preforms on new unseen data. The model then might work great on the training data, but not so great on new unseen data. If we split the dataset into train, validation and test, we can train the model on the training set, tune the model using the validation set and check the final performance on the test set. Here the training set still must be large (around $70\%$ of original dataset).
